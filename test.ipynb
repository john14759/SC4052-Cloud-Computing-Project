```python
from langchain import hub
from langchain.agents import AgentExecutor, create_react_agent, load_tools
from langchain_openai import AzureChatOpenAI
from langchain_community.utilities import ArxivAPIWrapper
import os
from langchain.prompts import PromptTemplate
from langchain.tools import Tool

llm = AzureChatOpenAI(
    azure_endpoint=os.environ['AZURE_OPENAI_ENDPOINT'],
    api_key=os.environ['AZURE_OPENAI_APIKEY'],
    deployment_name=os.environ['AZURE_OPENAI_DEPLOYMENT_NAME'],
    model_name=os.environ['AZURE_OPENAI_MODEL_NAME'],
    api_version=os.environ['AZURE_OPENAI_API_VERSION'],
    temperature=0
)

# Create the ArxivAPIWrapper instance
arxiv_wrapper = ArxivAPIWrapper(
    top_k_results=5,
    ARXIV_MAX_QUERY_LENGTH=300,
    load_max_docs=5,
    load_all_available_meta=False,
    doc_content_chars_max=40000
)

# Create a Tool with the ArxivAPIWrapper
arxiv_tool = Tool(
    name="arxiv",
    func=arxiv_wrapper.run,
    description="Useful for when you need to answer questions about scientific research papers. The input should be a search query."
)

tools = [arxiv_tool]
prompt = hub.pull("hwchase17/react")
agent = create_react_agent(llm, tools, prompt)
agent_executor = AgentExecutor(agent=agent, tools=tools, verbose=True, handle_parsing_errors=True)

response = agent_executor.invoke({
    "input": "Give me a list of research papers about Blockchain healthcare IOT",
})
```

```python
from langchain import hub
from langchain.agents import AgentExecutor, create_openai_functions_agent
from langchain_openai import AzureChatOpenAI
from langchain_community.tools.semanticscholar.tool import SemanticScholarQueryRun
import os

instructions = """You are an expert researcher."""
base_prompt = hub.pull("langchain-ai/openai-functions-template")
prompt = base_prompt.partial(instructions=instructions)

llm = AzureChatOpenAI(
    azure_endpoint=os.environ['AZURE_OPENAI_ENDPOINT'],
    api_key=os.environ['AZURE_OPENAI_APIKEY'],
    deployment_name=os.environ['AZURE_OPENAI_DEPLOYMENT_NAME'],
    model_name=os.environ['AZURE_OPENAI_MODEL_NAME'],
    api_version=os.environ['AZURE_OPENAI_API_VERSION'],
    temperature=0
)

tools = [SemanticScholarQueryRun()]

agent = create_openai_functions_agent(llm, tools, prompt)
agent_executor = AgentExecutor(
    agent=agent,
    tools=tools,
    verbose=True,
)

agent_executor.invoke(
    {
        "input": "Give me a list of research papers about Blockchain healthcare IOT."
        "List out the papers."
        "Break down the task into subtasks for search if needed. Use the search tool"
    }
)
```
